```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("store.csv")
df.head()
```

```python
df.groupby(["product", "revenue", "qty"]).max()
```

```python
# Calcular a mediana de revenue
threshold = df["revenue"].median()
```

```python
# Criar nova variável "consumers"
df["consumers"] = df["revenue"].apply(lambda x: "largest_reps" if x >= threshold else "reps_minors")
```

```python
# Salvar em um novo CSV
df.to_csv("store_with_consumers.csv", index=False)
```
